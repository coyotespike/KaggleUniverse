```python
# Handy shizzle
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## With the whole thing?!

```python
# Get the data and tidy it up
df = pd.read_csv('../input/train_1.csv')
df = df.drop(df.columns[3:], axis=1).fillna(0)
# Add spider variable
df['spider'] = 0
df.ix[df.Page.str.contains('spider'), 'spider'] = 1
df.head()
```

```python

# Have a punt!
dfLR = LinearRegression()
x = df.drop(['Page', '2015-07-02'], axis=1)
y = df['2015-07-02']
```

```python
dfLR.fit(x, y)
```

```python
dfLR.score(x, y)
```

```python
results = dfLR.predict(x)
results[0:10]
```

```python
results = [max(0, num) for num in results]
results[0:10]
```
